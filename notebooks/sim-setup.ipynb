```python
%config InlineBackend.figure_formats = {"retina", "png"}
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.size"] = 14

import tdgl
```

```python
points_per_cycle = 51
cycles = 1.5
npoints = int(points_per_cycle * cycles)
thetas = np.linspace(0, 2 * np.pi * cycles, npoints)
skip = int(0.5 * points_per_cycle)
I_fc_pks = np.linspace(0.5, 10, 201)
```

```python
fig, ax = plt.subplots()
ax.plot(
    thetas[0] / (2 * np.pi),
    np.cos(thetas)[0],
    "kX",
    label="$\\tau_\\mathrm{max}=2\\times10^4\\tau_0$",
)
ax.plot(
    thetas[1:skip] / (2 * np.pi),
    np.cos(thetas)[1:skip],
    "rX",
    label="Excluded from $M$ calc.\n$\\tau_\\mathrm{max}=2\\times10^3\\tau_0$"
)
ax.plot(
    thetas[skip:] / (2 * np.pi),
    np.cos(thetas)[skip:],
    "C0o",
    label="Included in $M$ calc.\n$\\tau_\\mathrm{max}=2\\times10^3\\tau_0$"
)

ax.grid(True)
ax.set_xticks([0, 0.5, 1, 1.5])
ax.set_yticks([-1, 0, 1])
ax.set_xlabel("$\\omega t/2\pi$")
ax.set_ylabel("$I_\\mathrm{FC}/I_{\\mathrm{FC},\\,\\mathrm{pk}}=\\cos(\\omega t)$")

ax2 = ax.secondary_xaxis(
    "top",
    functions=(lambda x: x * points_per_cycle, lambda x: x / points_per_cycle)
)
ax2.set_xlabel("Index")
ax2.set_xticks(np.arange(0, 90, 10))
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
fig.set_facecolor("white")

fig.savefig("../data/simulation/ac-cycle.png", dpi=200, bbox_inches="tight")
```
